## What is Isomorphism?

In Graph theory studies, two graphs are said to be isomorphic if they are structly the same.

![alt](./graph.png)

```python
import matplotlib.pyplot as plt

x = [1,3,3,1,1]
y = [3,3,1,1,3]

a = [4,4,6,6,4]
b = [3,1,3,1,3]

plt.ylim(0,6)
plt.xlim(0,8)

plt.plot(x, y, marker='o')
plt.plot(a, b, marker='o')
plt.show()
```
