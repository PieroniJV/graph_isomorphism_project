## What is Isomorphism?

In Graph theory studies, two graphs are said to be isomorphic if they are structly the same.

```python
import matplotlib.pyplot as plt

x = [1,3,3,1,1]
y = [3,3,1,1,3]

a = [4,4,6,6,4]
b = [3,1,3,1,3]

plt.ylim(0,4)
plt.xlim(0,7)

plt.plot(x, y, marker='o')
plt.plot(a, b, marker='o')
plt.show()
```

###Equally, if we "rotate" the edges of the graphs above, both graphs should still be isomorphic

```python
import matplotlib.pyplot as plt

x = [3,3,1,1,3]
y = [1,3,3,1,1]

a = [4,6,4,6,4]
b = [3,3,1,1,3]

plt.ylim(0,4)
plt.xlim(0,7)

plt.plot(x, y, marker='o')
plt.plot(a, b, marker='o')
plt.show()
```
