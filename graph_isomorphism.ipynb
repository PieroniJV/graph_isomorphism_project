## What is Isomorphism?
<p><a href="https://www.britannica.com/science/isomorphism-mathematics">Britannica.com</a> defines Isomorphism as follows: <q> A one-to-one correspondence between two sets that preserves binary relationships between elements of the sets. </q>

In simpler terms, isomorphism is a concept in mathematics that describes a relationship between two structures that have the same underlying shape or structure, but may differ in their specific elements or labels. Two objects are said to be isomorphic if they have the same structure, but may look different on the surface.

For example, two graphs with different vertex labels and edge weights may still be isomorphic if they have the same overall structure of connections between vertices.

Here follows an example to illustrate this:</p>

```python
x = [1,3,3,1,1]
y = [3,3,1,1,3]

a = [4,4,6,6,4]
b = [3,1,3,1,3]

plt.ylim(0,4)
plt.xlim(0,7)

plt.plot(x, y, marker='o')
plt.plot(a, b, marker='o')
plt.show()
```

Equally, if we "rotate" the edges of the graphs above, both graphs should still be isomorphic

```python
x = [3,3,1,1,3]
y = [1,3,3,1,1]

a = [4,6,4,6,4]
b = [3,3,1,1,3]

plt.ylim(0,4)
plt.xlim(0,7)

plt.plot(x, y, marker='o')
plt.plot(a, b, marker='o')
plt.show()
```

# Permutations

Permutations play an important role in graph isomorphism because they allow us to rearrange the vertices and edges of a graph while preserving its structure.

Some of the permutations that can be applied to graphs are:

<ul>

    <li>Edge permutation</li>

### Vertex permutation

This involves permuting the labels of the vertices in a graph. For example, consider the following two graphs:

```python
# Define Graph 1
G1 = nx.Graph()
G1.add_nodes_from(['A', 'B', 'C'])
G1.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'C')])

# Define Graph 2
G2 = nx.Graph()
G2.add_nodes_from(['D', 'E', 'F'])
G2.add_edges_from([('D', 'E'), ('E', 'F'), ('F', 'D')])

# Draw Graph 1
nx.draw(G1, with_labels=True)
plt.show()

# Draw Graph 2
nx.draw(G2, with_labels=True)
plt.show()

# Check if the graphs are isomorphic
print(nx.is_isomorphic(G1, G2))
```

These two graphs are isomorphic, because they have the same structure.

If we apply a vertex permutation to graph 2, we can see that they are still isomorphic.

```python
v_perm = {'D': 'A', 'E': 'B', 'F': 'C'}

# Apply the vertex permutation to Graph 2
G2_perm = nx.relabel_nodes(G2, v_perm)

# Check if the graphs are isomorphic after the vertex permutation
print(nx.is_isomorphic(G1, G2_perm))
```

Similarly, if we apply an edge permutation to graph 2, we can see that they are still isomorphic.

```python
# Define the edge permutation
e_perm = {('D', 'E'): ('A', 'B'), ('E', 'F'): ('B', 'C'), ('F', 'D'): ('C', 'A')}

# Apply the edge permutation to Graph 2
G2_perm = nx.relabel_nodes(G2, {'D': 'A', 'E': 'B', 'F': 'C'})
G2_perm = nx.relabel_nodes(G2_perm, e_perm)

# Check if the graphs are isomorphic after the edge permutation
print(nx.is_isomorphic(G1, G2_perm))
```

Lets see a different example now:

```python
# Define Graph 1
G3 = nx.Graph()
G3.add_nodes_from(['A', 'B', 'C'])
G3.add_edges_from([('A', 'B'), ('B', 'C')])

# Define Graph 2
G4 = nx.Graph()
G4.add_nodes_from(['D', 'E', 'F'])
G4.add_edges_from([('D', 'E'), ('E', 'F'), ('F', 'D'), ('D', 'F')])

# Draw Graph 1
plt.subplot(121)
nx.draw(G3, with_labels=True)
plt.title('Graph 1')

# Draw Graph 2
plt.subplot(122)
nx.draw(G4, with_labels=True)
plt.title('Graph 2')

plt.show()

# Check if the graphs are isomorphic
print(nx.is_isomorphic(G3, G4))
```

<p>In this example, Graph 1 has nodes A, B, and C, and edges (A, B) and (B, C). Graph 2 has nodes D, E, and F, and edges (D, E), (E, F), (F, D), and (D, F). 

When we plot the two graphs using the draw() function, we can see that they have different structures. 

Finally, we check that the graphs are not isomorphic by using the is_isomorphic() function, which should return 'False'.</p>

It is not possible to make the two graphs isomorphic to each other by any permutation of their nodes or edges.

Graph 1 has a degree sequence of [2, 2, 1], while Graph 2 has a degree sequence of [2, 2, 2]. 

The [degree sequence](#degree-sequence) is a graph invariant, meaning that it does not change under any graph isomorphism. Therefore, since the degree sequences of the two graphs are different, they cannot be isomorphic to each other.

## The Graph Isomorphism Problem
<hr>

The graph isomorphism problem is a computational problem in graph theory. This problem asks whether two given graphs are isomorphic, which means that they have the same structure despite potentially having different node and edge labels.

The "problem" is not known to be solvable in [polynomial time](#polynomial) nor to be [NP-complete](#np-complete). At the same time, isomorphism for many special classes of graphs can be solved in polynomial time, and in practice graph isomorphism can often be solved efficiently.

This problem has important applications in computer science, particularly in areas such as cryptography, network analysis, and computational chemistry. However, despite being studied for several decades, the graph isomorphism problem is still considered a challenging problem, and it is not known whether it can be solved in polynomial time.

The following example compares two undirected graphs G1 and G2. The graphs are represented as instances of the nx.Graph() class, and edges are added to the graphs using the add_edges_from() method.

The nx.is_isomorphic() method is used to check if the two graphs are isomorphic, which means that they have the same structure but may have different node and edge labels. If the graphs are isomorphic, the is_isomorphic variable will be set to True, otherwise, it will be set to False.

Finally, the program prints a message indicating whether the two graphs are isomorphic or not, depending on the value of the is_isomorphic variable.

```python
import networkx as nx

# define two graphs
G1 = nx.Graph()
G1.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4)])

G2 = nx.Graph()
G2.add_edges_from([(4, 3), (4, 2), (3, 2), (2, 1)])

# check if the two graphs are isomorphic
isomorphic = nx.is_isomorphic(G1, G2)

if isomorphic:
    print("The two graphs are isomorphic.")
else:
    print("The two graphs are not isomorphic.")
```

### Graph Isomorphism in Cryptography
<hr>

One application of graph isomorphism in criptography it is such as in the design and analysis of cryptographic protocols and algorithms.

Here's an example of how graph isomorphism can be used in cryptography:

Suppose you have two graphs G1 and G2, and you want to encrypt a message using one of these graphs. You can encrypt the message by assigning each letter a node in the graph and then replacing the letter with the corresponding node. To decrypt the message, you need to know which graph was used to encrypt it. If you know that the two graphs are structurally identical, then you can decrypt the message by using the same graph to replace the nodes with the corresponding letters.

```python
import random

# Define the graphs G1 and G2
G1 = {
    'A': ['B', 'C', 'D'],
    'B': ['A', 'C', 'E'],
    'C': ['A', 'B', 'D', 'E'],
    'D': ['A', 'C'],
    'E': ['B', 'C']
}

G2 = {
    'A': ['B', 'D'],
    'B': ['A', 'C', 'D', 'E'],
    'C': ['B', 'E'],
    'D': ['A', 'B', 'E'],
    'E': ['B', 'C', 'D']
}

# Define a message to be encrypted
message = 'HELLO WORLD'

# Choose one of the graphs randomly to encrypt the message
graph = random.choice([G1, G2])

# Encrypt the message by replacing each letter with its corresponding node in the graph
encrypted_message = ''
for letter in message:
    if letter in graph:
        encrypted_message += random.choice(graph[letter])
    else:
        encrypted_message += letter

# Print the encrypted message
print('Encrypted message:', encrypted_message)

# Decrypt the message using the same graph that was used to encrypt it
decrypted_message = ''
for character in encrypted_message:
    if character in graph.values():
        for letter, nodes in graph.items():
            if character in nodes:
                decrypted_message += letter
                break
    else:
        decrypted_message += character

# Print the decrypted message
print('Decrypted message:', decrypted_message)
```

This code defines two graphs, G1 and G2, and chooses one of them randomly to encrypt the message. The message is then encrypted by replacing each letter with a random node in the chosen graph. To decrypt the message, the same graph is used to replace the nodes with their corresponding letters. If the wrong graph is used to decrypt the message, the decrypted message will be gibberish.

### Graph Isomorphism in Network Analisis
<hr>

Another application of the graph isomorphism problem is in network analysis, which is the study of complex networks such as social networks, communication networks, biological networks, and many others. By identifying isomorphic subgraphs in a network, researchers can gain insight into the underlying structure and organization of the network, which can help them understand its behavior and properties.

For example, let's say we have a social network consisting of a set of individuals and their relationships with each other. We can represent this network as a graph, where each individual is a node, and an edge between two nodes represents a relationship between those individuals. If we want to study the structure of this social network, we may want to identify subgraphs that are isomorphic to each other. This could help us understand whether there are distinct groups or communities within the network, or whether certain individuals are more central or important than others.

In practice, network analysis often involves using algorithms to identify isomorphic subgraphs within a larger graph. For example, the VF2 algorithm is a well-known algorithm for solving the graph isomorphism problem, and it can be used to identify isomorphic subgraphs within a network. Other algorithms, such as subgraph isomorphism algorithms, can also be used for this purpose.

Overall, the graph isomorphism problem allows researchers to identify important subgraphs within a larger network, which can help them gain insight into the structure and behavior of the network.

Here is a simple, but practical example of the application of the graph isomorphism problem in network analisis. 

```python
import networkx as nx

# Create a sample graph representing a social network
G = nx.Graph()
G.add_edge('Alice', 'Bob')
G.add_edge('Bob', 'Charlie')
G.add_edge('Charlie', 'Denise')
G.add_edge('Denise', 'Eve')
G.add_edge('Eve', 'Frank')

# Identify isomorphic subgraphs using VF2 algorithm
GM = nx.algorithms.isomorphism.GraphMatcher(G, G)
for subgraph in GM.subgraph_isomorphisms_iter():
    print(subgraph)
```

In this example, we first create a simple graph representing a social network with five nodes and five edges. We then use the GraphMatcher function from the networkx library to find all subgraphs in G that are isomorphic to G itself. The subgraph_isomorphisms_iter function returns an iterator of all isomorphic subgraphs found, which we then print to the console.

### Graph Isomorphism Problem in Computational Chemistry
<hr>

Another important application of the graph isomorphism problem is in computational chemistry, where it is used to compare molecular structures and identify molecules that have similar properties. By identifying isomorphic subgraphs in a molecule, researchers can gain insight into its properties and behavior, which can help them design new drugs or materials.

For example, let's say we have two molecules, A and B, and we want to compare their structures to determine whether they are similar. We can represent each molecule as a graph, where atoms are nodes and bonds between atoms are edges. We can then use the graph isomorphism problem to determine whether there is a mapping between the nodes of A and the nodes of B that preserves the bonding relationships.

In practice, this often involves using algorithms to solve the graph isomorphism problem and identify subgraphs that are isomorphic to each other. For example, the Weisfeiler-Lehman algorithm is a well-known algorithm for solving the graph isomorphism problem, and it has been used in computational chemistry to compare molecular structures.

Overall, the graph isomorphism problem allows researchers to compare molecular structures and identify molecules that have similar properties. This can help in the design of new drugs or materials, as well as in understanding the behavior of existing compounds.

The following example how to represent a molecule as a graph and compare its structure to that of another molecule: 

```python
import networkx as nx

# Define the structure of molecule A
A = nx.Graph()
A.add_node(1, element='C')
A.add_node(2, element='O')
A.add_node(3, element='H')
A.add_edge(1, 2, order=2)
A.add_edge(1, 3, order=1)

# Define the structure of molecule B
B = nx.Graph()
B.add_node(1, element='C')
B.add_node(2, element='O')
B.add_node(3, element='H')
B.add_edge(1, 2, order=1)
B.add_edge(1, 3, order=1)

# Compare the structures of A and B
GM = nx.algorithms.isomorphism.GraphMatcher(A, B)
if GM.is_isomorphic():
    print("Molecules A and B have the same structure.")
else:
    print("Molecules A and B have different structures.")
```

In this example, we define the structures of two molecules, A and B, using the networkx library. We represent each molecule as a graph, where atoms are nodes and bonds between atoms are edges. We then use the GraphMatcher function from networkx to compare the structures of A and B.

The is_isomorphic function returns True if there is a mapping between the nodes of A and the nodes of B that preserves the bonding relationships, indicating that the two molecules have the same structure. If there is no such mapping, the function returns False, indicating that the two molecules have different structures.

In practice, comparing the structures of complex molecules may require more sophisticated algorithms and techniques than the simple example shown here. However, the basic approach of representing a molecule as a graph and using the graph isomorphism problem to compare its structure to that of another molecule is a common technique in computational chemistry.
